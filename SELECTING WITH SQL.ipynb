```python
#import necessary libraries
import pandas as pd
import sqlite3
#use the case statement to dynamically assign the credit limit of each customer
#creating connection/link to database
conn=sqlite3.connect('data.sqlite')
#creating a
my_df=pd.read_sql("""
                  SELECT *
                  FROM customers
                  ;""",conn)
my_df
```

```python
#use the case statement to dynamically assign the credit limit of each customer
# pd.read_sql("""
#             SELECT customerName,contactLastName,contactFirstName,
#             CASE
#                 WHEN
#             ;""",conn)
#converting columns to int objects and Handle non-numeric values
my_df['creditLimit'] = pd.to_numeric(my_df['creditLimit'], errors='coerce').astype('Int64')  # Nullable integer type
#finding the range between highest and lowest to create classes
print(my_df['creditLimit'].max())
print(my_df['creditLimit'].min())
range=my_df['creditLimit'].max()-my_df['creditLimit'].min()
print(range)

```

```python
#use the case statement to dynamically assign the credit limit of each customer
#converting columns to int objects and Handle non-numeric values
pd.read_sql("""
            SELECT contactLastName, contactFirstName, creditLimit,
                CASE 
                    WHEN CAST(creditLimit AS INTEGER) BETWEEN 0 AND 60000 THEN 'MINIMAL DEBT'
                    WHEN CAST(creditLimit AS INTEGER) BETWEEN 61000 AND 120000 THEN 'NORMAL DEBT'
                    WHEN CAST(creditLimit AS INTEGER) BETWEEN 121000 AND 180000 THEN 'RISKY DEBT'
                    ELSE 'CRITICAL DEBT'
                END AS debt_category
            FROM customers;
            """,conn).head(20)
```

```python
pd.read_sql("""
            SELECT contactFirstName,
            CASE
                WHEN LENGTH(contactFirstName) <= 4 THEN 'SHORT_NAME'
                WHEN LENGTH(contactFirstName) >=5 AND LENGTH(contactFirstName) < 7 THEN 'MEDIUM_LENGTH'
                ELSE 'LONG_NAME'
            END AS length_of_names
            FROM customers;""",conn).head(20)
```

```python
pd.read_sql("""
            SELECT UPPER(contactFirstName) AS all_caps_column
            FROM customers;
            """,conn)
```

```python
pd.read_sql("""
            SELECT CONCAT(contactFirstName,' ',contactLastName) AS FULLNAME,
            CASE
            WHEN CONCAT(contactFirstName,' ',contactLastName) LIKE '%Sue Taylor' THEN 'IT WORKS'
            END AS 'COMPLETE'
            FROM customers
            ;""",conn)
```
